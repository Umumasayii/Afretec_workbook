```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("Experience-Salary.csv")
df.head()
```

```python
x = df["exp(in months)"]
y = df["salary(in thousands)"]
```

```python
df.isnull().sum()
```

```python
sns.boxplot([x,y])
```

$$
\text{Manual Calculations Of Linear Model}
$$

$$
Y = B_0 + B_1X \\
\text{Where } \, B_1 =\frac{\frac{1}{n} \sum(X_iY_i)-\mu_x\mu_y}{\frac{1}{n} \sum(X_i^2) - \mu_x^2} \\
\text{And }\, B_0 = \mu_y - B_1\mu_x
$$

```python
n = len(x)

b1 = (np.sum(x*y) - n*np.mean(x)*np.mean(y)) / \
     np.sum((x - np.mean(x))**2)
b1
```

```python
b0 = np.mean(y)-b1*np.mean(x)
b0
```

$$
\text{Y = 5.2+0.8X}
$$

```python
beta, beta0, r_value, p_value, std_err = stats.linregress(x,y)
print("y = %f x + %f, r: %f, r-squared: %f,\np-value: %f, std_err: %f" % (beta, beta0, r_value, r_value**2, p_value, std_err)) 
print("Regression line with the scatterplot") 
yhat = beta * x + beta0 
plt.plot(x, yhat, 'r-', x, y,'o')
plt.title("The Relationship Between Samary and Experience", loc = "center") 
plt.xlabel('Experience (years)') 
plt.ylabel('Salary') 
plt.show()
```
